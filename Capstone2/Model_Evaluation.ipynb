## Using RMSE to Evaluate Model

```python
import pandas as pd
import numpy as np
```

```python
# loading 2019-20 projections
proj_df = pd.read_csv(r"C:\Users\15127\Springboard\Capstone2\nba-stats-csv\proj_2019.csv", header = 0)
```

```python
proj_df.columns
```

```python
# renaming columns to match with real_df
proj_df.rename(columns = {'proj_season_id':'season_id',
                               'proj_pts': 'pts',
                                'proj_min':'min',
                                'proj_fgm':'fgm',
                                'proj_fga':'fga',
                                'proj_fg3m':'fg3m',
                                'proj_fg3a':'fg3a',
                                'proj_ftm':'ftm',
                                'proj_fta':'fta',
                                'proj_oreb':'oreb',
                                'proj_dreb':'dreb',
                                'proj_ast':'ast',
                                'proj_stl':'stl',
                                'proj_tov':'tov',
                               'proj_blk':'blk'}, inplace = True)
```

```python
proj_df = proj_df.dropna(how='all')
proj_df.drop('Unnamed: 0',axis = 1,inplace=True)
proj_df.reset_index(drop=True,inplace=True)
```

```python
proj_df
```

```python
# loading actual results of 2019-20 season
real_df = pd.read_csv(r"C:\Users\15127\Springboard\Capstone2\nba-stats-csv\NBA_Stats.csv", header = 0)
```

```python
# cleaning...
real_df = real_df.dropna(how='all')
real_df.drop('Unnamed: 0',axis = 1,inplace=True)
real_df.reset_index(drop=True,inplace=True)
```

```python
# isolating 2019-20 season
real_df = real_df[real_df['SEASON'] == '2019-20']
```

```python
# aligning columns with proj_df
real_df = real_df[['PLAYER_ID','SEASON','PTS', 'GP','AGE','MIN','FGM','FGA','FG_PCT',
                   'FG3M','FG3A','FG3_PCT','FTM','FTA','FT_PCT','OREB','DREB',
                  'AST', 'TOV','STL','BLK']]
```

```python
# matching names
real_df.columns = ['player_id','season_id','pts','gp','age','min','fgm','fga','fg_pct',
                   'fg3m','fg3a','fg3_pct','ftm','fta','ft_pct','oreb','dreb',
                   'ast', 'tov', 'stl','blk']
```

```python
real_df.head()
```

```python
# merging data
merged = proj_df.merge(real_df,how='left',on=['player_id','season_id'])
```

```python
# dropping any rows with NaN values
merged = merged.dropna(how='any')
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# isolating real stats
real_df = merged.loc[:, ['pts_y','min_y','fgm_y','fga_y','fg3m_y','fg3a_y','ftm_y','fta_y','oreb_y','dreb_y','ast_y','stl_y','tov_y', 'blk_y']]
```

```python
# isolating projected stats
proj_df = merged.loc[:, ['pts_x','min_x','fgm_x','fga_x','fg3m_x','fg3a_x','ftm_x','fta_x','oreb_x','dreb_x','ast_x','stl_x','tov_x', 'blk_x']]
```

```python
real_df
```

```python
proj_df
```

```python
# RMSE
se = mean_squared_error(real_df, proj_df)
```

```python
# taking average
mse = np.mean(se)
```

```python
# squaring
rmse = np.sqrt(mse)
```

```python
confidence = rmse
```

```python
print('{0} percent confidence in projected {1} per game stats'.format(100 - round(confidence, 2), '2019-20'))
```
